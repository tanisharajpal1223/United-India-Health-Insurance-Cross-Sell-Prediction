## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Create a shallow copy of the original dataframe
ins_df = df.copy()

# Drop the id column from the data frame
ins_df.drop('id', inplace = True, axis = 1)
````

````python
# Convert columns with binary values into categorical values
ins_df['Driving_License'] = ins_df['Driving_License'].map({1:'Yes', 0:'No'})
ins_df['Previously_Insured'] = ins_df['Previously_Insured'].map({1:'Yes', 0:'No'})
ins_df['Response'] = ins_df['Response'].map({1:'Yes', 0:'No'})
````

````python
# Create Vintage category
def vin_category(vintage):
  if (vintage <= 100):
    return "Short-term Customers"
  if (vintage > 100 and vintage <= 200):
    return "Intermediate Customers"
  else:
    return "Long-term customers"

ins_df['Vintage Category'] = ins_df['Vintage'].apply(vin_category)

````

````python
# Show the data frame after converting
ins_df.head()
````

````python
# Convert float data types to int data types.
ins_df[['Region_Code','Annual_Premium', 'Policy_Sales_Channel']] = ins_df[['Region_Code','Annual_Premium', 'Policy_Sales_Channel']].astype(int)
````

### What all manipulations have you done and insights you found?

Answer Here.
1. We created a shallow copy of the original dataframe for backup purposes
2. We dropped the unique identifier from the dataframe as it does not contribute to our visualization.
3. In this step, we found that there are three columns, "Driving License",Previously_Insured, and Response which contain binary values. These should be converted into categorical values: "Yes" and "No". We assigned "Yes" to 1 and "No" to 0. Converting the binary values into categorical values will ease the understanding of visualizations.
4. We converted the float data types into integers to reduce memory usage.
5. We created an additional column based on the "vintage" column to categorize the types of customers in our dataset, distinguishing between long-term, intermediate, and short-term customers.

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.
1. The average age of the company's customers is greater than or equal to 38.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.

The average age of the company's customers is greater than or equal to 38.

Null Hypothesis (H0): µ >=  38

Alternate Hypothesis (Ha): µ < 38

#### 2. Perform an appropriate statistical test.

````python

# Perform Statistical Test to obtain P-Value


random_seed = 42
np.random.seed(random_seed)

age_df_sample = ins_df['Age'].sample(300)
age_df_sample_mean = np.mean(age_df_sample)
population_std = np.std(ins_df['Age'])

population_age_mean = 38
print("The mean age of the customers of the company:", population_age_mean)
print("The mean age of the 300 samples:",age_df_sample_mean)
print("Standard deviation of the customers of the company:", population_std)

# Define the parameters for z test
n = 300
mu = population_age_mean
x_bar = age_df_sample_mean
sigma = population_std

alpha = 0.05

# Obtain p value
pvalue = norm.cdf(x_bar, loc = mu, scale = (sigma / np.sqrt(n)))
print("The p value obtained is:", pvalue)


print('Significance level of left side:', alpha)

# Left tail test
if pvalue < alpha:
  print("Reject H0")
else:
  print("Fail to reject H0")
````

##### Which statistical test have you done to obtain P-Value?

Answer Here.

I have used Z-test to obtain p value. After conducting a left-tailed Z-test with a 95% confidence level, we obtained a p-value greater than significance level 0.05. Consequently, we failed to reject the null hypothesis. This suggests that the mean age of 37 years, calculated from a random sample of 300 individuals, falls within the acceptance region and is not significantly different from the population mean. Hence the claim made by the company is true.

##### Why did you choose the specific statistical test?

Answer Here.

I chose the Z-test because it's suitable for comparing means when you have a relatively large sample size, like the 300 samples. It's commonly used when we know the population standard deviation or when the sample size is large enough for the sample standard deviation to be a good estimate of the population standard deviation. Since we were comparing a sample mean to a population mean, the Z-test was the right choice.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.

The probability of customers who were previously insured by the company is greater than or equal to 50%.

Null Hypothesis (H0): p >=  0.5

Alternate Hypothesis (Ha): p < 0.5

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value


random_seed = 42
np.random.seed(random_seed)


sample_ins = ins_df.sample(500)
prob_ins_sample = len(sample_ins[sample_ins['Previously_Insured'] == "Yes"]) / len(sample_ins)

population_prob = 0.5
print("Probability of the sample is:",prob_ins_sample)
print("Probability of the population is:",population_prob)


# Define the parameters for z test
n = 500
p_cap = prob_ins_sample
p = population_prob
q = 1-p
alpha = 0.05
print('Significance level of left side:', alpha)

# Obtain p value
pval = norm.cdf(p_cap, loc = p, scale = np.sqrt(p * q / n))
print("P value is:", pval)


# Left tail test
if pval < alpha:
  print("Reject H0")
else:
  print("Fail to reject H0")
````
